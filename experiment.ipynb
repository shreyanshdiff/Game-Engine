```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
```

```python
def safe_str(x):
    if pd.isna(x):
        return ''
    return str(x)
```

```python
df['features'] = df.apply(lambda row: 
    ' '.join(safe_str(x) for x in row['Popular Tags']) + ' ' + 
    ' '.join(safe_str(x) for x in row['Game Features']) + ' ' +
    safe_str(row['Game Description']) + ' ' +
    safe_str(row['Title']), axis=1)
```

```python
tfidf = TfidfVectorizer(stop_words='english' , max_features=5000)
sparse_matrix = tfidf.fit_transform(df['features'])
```

```python
knn = NearestNeighbors(n_neighbors=6, metric='cosine', algorithm='brute')
knn.fit(sparse_matrix)
```

```python
def recommend_games_memopt(title, data=df, model=knn, top_n=5):
    try:
        idx = data[data['Title'] == title].index[0]
        query = sparse_matrix[idx]
        
        distances, indices = model.kneighbors(query, n_neighbors=top_n+1)
        
        results = []
        for i, distance in zip(indices[0][1:], distances[0][1:]):  # Skip self
            results.append({
                'Title': data.iloc[i]['Title'],
                'Link': data.iloc[i]['Link'],
                'Similarity': f"{1 - distance:.2f}",  # Convert distance to similarity
                'Price': data.iloc[i].get('Discounted Price', 'N/A')
            })
            
        return pd.DataFrame(results)
    
    except Exception as e:
        print(f"Error: {e}")
        return pd.DataFrame()
```

```python
recommendations = recommend_games_memopt("Winter Memories")
print(recommendations)
```
